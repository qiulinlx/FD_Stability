```python
import matplotlib.pyplot as plt
import numpy as np

bins_gc = np.linspace(sd_df['GC'].min(), sd_df['GC'].max(), 11)  # 11 edges = 10 bins
labels = [f'{bins_gc[i]:.2f}–{bins_gc[i+1]:.2f}' for i in range(10)]

sd_df['GC_bin'] = pd.cut(sd_df['GC'], bins=bins_gc, labels=labels, include_lowest=True)

cols = 5
rows = 2
fig, axes = plt.subplots(rows, cols, figsize=(cols * 12, rows * 10))
axes = axes.flatten()

for i, label in enumerate(labels):
    subset = sd_df[sd_df['GC_bin'] == label]
    axes[i].hist2d(subset["Simpson's Index"], subset['transformed npp'], bins=30, cmap='viridis')
    axes[i].set_title(f'GC: {label}')
    axes[i].set_xlabel('Gini Coefficient of DBH')
    axes[i].set_ylabel('Transformed NPP')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

biomes = sd_df['biome_y'].dropna().unique()
n = len(biomes)
cols = 3
rows = math.ceil(n / cols)

fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4))
axes = axes.flatten()

for i, biome in enumerate(biomes):
    subset = sd_df[sd_df['biome_y'] == biome]
    axes[i].hist2d(subset['GC'], subset['transformed npp'], bins=30, cmap='viridis')
    axes[i].set_title(biome)
    axes[i].set_xlabel('Gini Coefficient of Diameter at Breast Height')
    axes[i].set_ylabel('transformed npp')

for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```
